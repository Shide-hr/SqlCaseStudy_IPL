```python
%%sql
-- total spending on player for each team:

SELECT Team, SUM(Price_in_cr) as "total_spending"  FROM IPLPlayers
GROUP BY Team
ORDER BY total_spending DESC;
```

```python
%%sql
-- The top 3 highest-paid 'All-rounder' across all teams:

SELECT Player, Price_in_cr, Role FROM IPLPlayers 
WHERE Role = 'All-rounder'
ORDER BY Price_in_cr DESC
LIMIT 3;
```

```python
%%sql
-- Highest_price player in each team:

SELECT Player, MAX(Price_in_cr) as 'Highest_price' , Team From IPLPlayers
GROUP BY Team
ORDER BY Price_in_cr DESC;
```

```python
%%sql

-- Rank players by their price within the team and list the top 2 of each team:

WITH RANKED_PLAYERS AS(
SELECT Player, Price_in_cr, Team,
ROW_NUMBER() OVER(PARTITION BY Team ORDER BY Price_in_cr DESC) AS 'Rank_Within_Team'
FROM IPLPlayers)

SELECT Player, Price_in_cr, Team, Rank_Within_Team
FROM RANKED_PLAYERS
WHERE Rank_Within_Team <= 2
;
```

```python
%%sql
-- the most expensive player of each team, along with the second most expensive player´s name and price:

WITH RANKED_PLAYER AS(
    
SELECT Player, Team, Price_in_cr,
ROW_NUMBER() OVER (
    PARTITION BY Team
    ORDER BY Price_in_cr DESC) AS 'RANK_OF_PLAYER'
    
FROM IPLPlayers)

SELECT Team,
MAX(CASE WHEN RANK_OF_PLAYER = 1 THEN Player END) AS 'Most_expensive_player',
MAX(CASE WHEN RANK_OF_PLAYER = 1 THEN Price_in_cr END) AS 'Highest_price',
MAX(CASE WHEN RANK_OF_PLAYER = 2 THEN Player END) AS 'Second_most_expensive_player',
MAX(CASE WHEN RANK_OF_PLAYER = 2 THEN Price_in_cr END) AS 'Second_highest_price'


FROM RANKED_PLAYER
GROUP BY Team;
```

```python
%%sql
-- percentage of each player´s price to their teams total spending

SELECT Player, Team,  Price_in_cr,
 ROUND((Price_in_cr / SUM( Price_in_cr) OVER (PARTITION BY Team)) * 100 ,2) AS 'Contribution_percentage'

FROM IPLPlayers;
```

```python
%%sql
-- classify the players as 'high', 'medium' or 'low' price based on these rules:
-- high : price > 15 cr
-- medium : price between 5 cr and 15 cr
-- low : price < 5

WITH CR_CL AS (
    SELECT Player, Team, Price_in_cr,
        CASE 
            WHEN Price_in_cr > 15 THEN 'HIGH'
            WHEN Price_in_cr BETWEEN 5 AND 15 THEN 'MEDIUM'
            WHEN Price_in_cr < 5 THEN 'LOW'
    
        END AS 'Price_category'
    
    FROM IPLPlayers
    )

SELECT Team , Price_category , Count(*) as 'Number_Of_Players'


FROM CR_CL
GROUP BY Team, Price_category
;
```

```python
%%sql
-- Average price of indian players and compar it with overseas players

SELECT 
    CASE
        WHEN Type LIKE 'Indian%' THEN 'Indian'
        WHEN Type LIKE 'Overseas%' THEN 'Overseas'

    END AS 'Player_type', 
ROUND (AVG (Price_in_cr),2) AS 'AVG_price'


FROM IPLPlayers

GROUP BY Player_type
;
```

```python
%%sql
-- players who earns more than the average price of their team

WITH PL_CT AS(
    SELECT  Team, Player , Price_in_cr,
    AVG(Price_in_cr) OVER (PARTITION BY Team) AS 'AVG_Price_of_Team'
    
    
    FROM IPLPlayers)

SELECT Player, Price_in_cr, AVG_Price_of_Team, Team,
    CASE
        WHEN Price_in_cr > AVG_Price_of_Team THEN 'YES'
        ELSE 'NO'

    END  AS 'IF_MORE_THAN_AVG'

FROM PL_CT

WHERE IF_MORE_THAN_AVG LIKE 'YES' ;
```

```python
%%sql
-- the most expensive player in each role

WITH MA_CT AS (
    SELECT Player, Price_in_cr, Role, 
    
    max(Price_in_cr) OVER (PARTITION BY Role) AS 'Max_price'
    
    FROM IPLPlayers
)

SELECT Player , Role, Max_price

FROM MA_CT

WHERE Price_in_cr = Max_price

;
```
